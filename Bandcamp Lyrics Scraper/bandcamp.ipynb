```python
from bs4 import BeautifulSoup as bs
from lxml import html
import requests as r
```

```python
# URL part. Later will ask for input
url='https://galleonsmusic.bandcamp.com/album/metropolis'
```

```python
# Get page
rp=r.get(url)
```

```python
# Create bs object
soup=bs(rp.text, 'lxml')
print(soup.prettify())
```

```python
# Export entire html tree to a file for viewing (Comment this out later)
with open('htmlstring','w', encoding='utf-8',) as file1:
    file1.write(soup.prettify())
file1.close()
```

```python
# Extract the meta data for album
album_data= soup.find('table', class_="track_list track_table", id="track_table")
print(album_data)
```

```python
# From meta data extract the song title number, song title, and song length
# Starting with title number
title_number=album_data.find('div', class_='track_number secondaryText').text.strip()
title_number
```

```python
# Song title
song_title=album_data.find('span', class_='track-title').text.strip()
song_title
```

```python
# Song length
song_length=album_data.find('span', class_='time secondaryText').text.strip()
song_length
```
